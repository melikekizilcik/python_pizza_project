```python
import csv
import datetime

# Menu.txt oluşturuldu
with open("Menu.txt","r") as menu:
   for menu in menu:
        print(menu)

# "Pizza" üst sınıfı tanımlandı
class Pizza:
   def __init__(self, description, cost): 
        self.description = description     
        self.cost = cost 

   def get_description(self):
        return self.description 

   def get_cost(self):
        return self.cost 

# "Pizza" alt sınıfları
class Klasik(Pizza):
    def __init__(self):
        super().__init__("Klasik pizza", 69.99)

class Margherita(Pizza):
    def __init__(self):
        super().__init__("Margherita", 69.99)

class TurkishPizza(Pizza):
    def __init__(self):
        super().__init__("Turkish pizza", 85.00)

class DominosPizza(Pizza):
    def __init__(self):
        super().__init__("Dominos pizza", 79.99)

# Decorator sınıfı oluştur
class Decorator(Pizza):
  def __init__(self, component, description, cost):
        self.component = component
        self.description  = description
        self.cost = cost

  def get_cost(self):
       return self.component.get_cost() + \
         Pizza.get_cost(self)


  def get_description(self):
       return self.component.get_description() + \
         ' ' + Pizza.get_description(self)


# Ekstra Malzeme sınıfları
class Olive(Decorator):
  def __init__(self, component):
    super().__init__(component, "Zeytin", 3)

class Mushroom(Decorator):
  def __init__(self, component):
    super().__init__(component, "Mantar", 3)

class Cheese(Decorator):
  def __init__(self, component):
    super().__init__(component, "Keçi Peyniri", 3)

class Meat(Decorator):
  def __init__(self, component):
    super().__init__(component, "Et", 3)

class Onion(Decorator):
  def __init__(self, component):
   super().__init__(component, "Soğan", 3)

class SweetCorn(Decorator):
  def __init__(self, component):
   super().__init__(component, "Mısır", 3)


# Menu.txt ekrana yazdırıldı
def main():
    with open("Menu.txt") as menu:
        for l in menu:
            print(l, end="")

    pizzaMap = {1: Klasik, 
           2: Margherita, 
           3: TurkishPizza, 
           4: DominosPizza,
           11: Olive, 
           12: Mushroom, 
           13: Cheese, 
           14: Meat, 
           15: Onion, 
           16: SweetCorn
              }

#Sipariş alma

    pizza = input("Lütfen Pizzanızı Seçiniz: ")
    while pizza not in ["1", "2", "3", "4"]:
        pizza = input("Üzgünüm, bu çeşit pizzamız bulunmamaktadır, başka çeşitlerimizi denemek ister misiniz?")

    newPizza = pizzaMap[int(pizza)]()

    while pizza != "ok":
        pizza = input("Ekstra malzeme ister misiniz")
        if pizza in ["11", "12", "13", "14", "15", "16"]:
            newPizza = pizzaMap[int(pizza)](newPizza)

    print("\n"+newPizza.get_description().strip() +
          ": TL" + str(newPizza.get_cost()))
    print("\n")

#Kullanıcı bilgileri
    print("Sipariş oluşturuldu")
    firstName = input("Adınız: ")
    tcNo = input("TC Kimlik No: ")
    creditCard = input("Kredi Kartı Numaranız: ")
    creditCardPassword = input("Şifre: ")
    orderDate = datetime.datetime.now()

#Veritabanına ekleme
    with open('Orders_Database.csv', 'a') as orders:
        orders = csv.writer(orders, delimiter=',')
        orders.writerow([firstName, tcNo, creditCard, creditCardPassword, newPizza.get_description(), orderDate])
    print("Sipariş Onaylandı.")

#Projeyi çalıştırma
if __name__ == '__main__':
    main()
```
